```python
# import libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

#Load your RFM data
df = pd.read_csv("rfm_table.csv")

# Scale the RFM features (required for clustering)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[['Recency', 'Frequency', 'Monetary']])

# Find the best number of clusters using Elbow Method
inertia = []
K = range(1, 11)  # Try from 1 to 10 groups

for k in K:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(scaled_data)
    inertia.append(model.inertia_)  # Lower is better

# Step 6: Plot the Elbow Graph
plt.figure(figsize=(8,5))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia (Distance inside cluster)')
plt.title('Elbow Method for Finding Best K')
plt.grid(True)
plt.show()

#conclusion
print('Based on the bend in the graph (the "elbow point"), I concluded that K = 4 (for example) is the most appropriate number of clusters for segmenting customers.')




```
