```python
# import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Load the file
df = pd.read_csv("rfm_with_clusters.csv")
df.head()

# Scatter Plot  Recency vs Monetary
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Recency', y='Monetary', hue='Segment', palette='Set2')
plt.title("Recency vs Spending (Segment-wise)")
plt.grid(True)
plt.tight_layout()
plt.show()
print("This graph helps identify which customers are buying recently and spending the most.")

# Scatter Plot  Frequency vs Monetary
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Frequency', y='Monetary', hue='Segment', palette='Set1')
plt.title("Frequency vs Spending (Segment-wise)")
plt.grid(True)
plt.tight_layout()
plt.show()
print("This chart shows that customers who purchase more frequently also tend to spend more.")


# Box Plot  Recency for each Segment
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Segment', y='Recency')
plt.title("Recency Distribution by Segment")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
print("This graph compares how recently each group of customers last made a purchase.")

# Box Plot  Frequency for each Segment
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Segment', y='Frequency')
plt.title("Frequency Distribution by Segment")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
print("This plot shows how often each segment makes purchases.")

# Step 7: Box Plot  Monetary for each Segment
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Segment', y='Monetary')
plt.title("Spending Distribution by Segment")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
print("This chart shows which segments spend the most money overall.")

from IPython.display import Markdown

Markdown("""
###  Summary Table Graph Understanding

|  **Graph**                        |  **Purpose**                                 |  **What to Look For**                         |
|------------------------------------|------------------------------------------------|-------------------------------------------------|
| **Recency vs Monetary (scatter)**  | See who buys recently and spends more          | **Top-left = recent + high spenders (best)**    |
| **Frequency vs Monetary (scatter)**| See who buys often and spends more             | **Top-right = loyal + high spenders**           |
| **Boxplot: Recency by Segment**    | Compare recent purchase activity per group     | **Lower box = more recent = active**            |
| **Boxplot: Frequency by Segment**  | Compare how often each group buys              | **Higher box = frequent buyers**                |
| **Boxplot: Monetary by Segment**   | Compare how much each group spends             | **Higher box = big spenders**                   |
""")




```
