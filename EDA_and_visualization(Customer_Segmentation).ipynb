```python
# step 1 : import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#step 2 : load datasets
df = pd.read_csv('Mall_Customers.csv')

#step 3 : summary statistics
df.describe()
df.info()

#step 4 : data visualization (gender)
sns.countplot(x='Genre', data=df)
plt.title("Gender Distribution")
plt.show()

# age histogram
sns.histplot(df['Age'], bins=15, kde=True)
plt.title("Age Distribution")
plt.show()

# Income vs Spending Score
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=df, hue='Genre')
plt.title("Annual Income vs Spending Score")
plt.show()

# Boxplot of spending by gender
sns.boxplot(x='Genre', y='Spending Score (1-100)', data=df)
plt.title("Spending by Gender")
plt.show()
```
