```python
# step 1 : import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#step 2 : load datasets
df = pd.read_csv('E-Commerce data.csv')

# basic view
print(df.shape)
print(df.columns)
print(df.head())

#Top 10 countries with most customers
plt.figure(figsize=(10,5))
df['Country'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries by Customer Count')
plt.ylabel('Number of Customers')
plt.xlabel('Country')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
print('The most customers are from the United Kingdom')

#Quantity distribution
plt.figure(figsize=(8,4))
sns.histplot(df['Quantity'], bins=50, kde=False, color='orange')
plt.title('Distribution of Quantity Purchased')
plt.xlabel('Quantity')
plt.ylabel('Count')
plt.show()
print('Most products are purchased in small quantities (1â€“10)')

#Unit Price distribution
plt.figure(figsize=(8,4))
sns.boxplot(x=df['UnitPrice'])
plt.title('Unit Price Distribution')
plt.xlabel('Unit Price')
plt.show()
print("Many products have very low prices; a few outliers are highly priced")






```
